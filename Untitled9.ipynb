```python
%%writefile research_agent.py
import requests
from newspaper import Article
from openai import OpenAI


SERP_API_KEY = "f55b15a85cbe7249c1fbdebfaa17f56fa5242d93bef9d7af82fe63d494453283"
OPENAI_API_KEY = "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"
client = OpenAI(api_key=OPENAI_API_KEY)

def search_web(query, num_results=5):
    url = f"https://serpapi.com/search.json?q={query}&api_key={SERP_API_KEY}"
    response = requests.get(url).json()
    results = []
    for item in response.get("organic_results", [])[:num_results]:
        results.append({
            "title": item.get("title"),
            "link": item.get("link"),
            "snippet": item.get("snippet")
        })
    return results

def extract_content(url):
    try:
        article = Article(url)
        article.download()
        article.parse()
        return {
            "title": article.title,
            "text": article.text,
            "publish_date": article.publish_date
        }
    except Exception as e:
        return {"error": str(e)}

def summarize_text(text, source_title, url):
    prompt = f"""
    Summarize the following article in 3-5 bullet points.
    Title: {source_title}
    URL: {url}
    Article: {text[:5000]}
    """

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.3
    )
    return response.choices[0].message.content

def research_agent(query):
    results = search_web(query)
    report = []
    for r in results:
        content = extract_content(r["link"])
        if "text" in content:
            summary = summarize_text(content["text"], content["title"], r["link"])
            report.append({
                "title": content["title"],
                "url": r["link"],
                "summary": summary,
                "publish_date": content.get("publish_date")
            })
    return report
```

```python

%%writefile app.py
import streamlit as st
from research_agent import research_agent

st.title("ðŸ”Ž AI Research Agent")
query = st.text_input("Enter your research topic:")

if st.button("Run Research"):
    report = research_agent(query)
    for item in report:
        st.subheader(item["title"])
        st.write(f"[Read more]({item['url']})")
        st.write(item["summary"])
```
